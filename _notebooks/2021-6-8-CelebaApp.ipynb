# **The GreatCeleb Facial Attribute App!**
> Here we present to you our facial attributes detection app, which predicts your various facial attributes as compared to the attributes of many celebrity faces!

```python
#Defining the basic IPython Widgets
path=Path('../input/celeba-model')
learn = load_learner(path/'export_new.pkl',cpu=True)
btn_upload = widgets.FileUpload()
btn_run = widgets.Button(description='Predict')
lbl_pred = widgets.Label()
out_pl = widgets.Output()
```

```python
#Defining a function to Predict
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img)
    pred,pred_idx,probs = learn.predict(img)
    pred_str = str(pred)
    probabilities=str(probs[pred_idx])
    lbl_pred.value = f'Prediction: {pred};  Probability: {probabilities}'

btn_run.on_click(on_click_classify)
```

```python
#For Vertical GUI:
VBox([widgets.Label('Upload Your Favourite Face!'), 
      btn_upload, btn_run, out_pl, lbl_pred])
```
